```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'data/Chicken_Wings.csv'

df = pd.read_csv(file_path,
                 skiprows=1, #* skip symbols, header is inferred as 2nd row
                 )

print(df.head())
```

```python
df.plot(title="Average Cost of Wing vs Quantity Purchased",
        x="Quantity",
        y="Avg Wing Cost",
        xlabel="Quantity of Wings",
        ylabel="Avg Price of 1 Wing (cents)"
        )
```

```python
min_awc = df['Avg Wing Cost'].min() #? minimum average wing cost
max_awc = df['Avg Wing Cost'].max()

#* lambda function to find row based on value in 'Avg Wing Cost' col == val
rowFinder = lambda val: df[df['Avg Wing Cost'] == val]

#* select values in "Quantity" column for "awc" == min_awc
print(f"Best deal on wings is for {min_awc} by purchasing {rowFinder(min_awc)['Quantity'].tolist()} wing(s)")
print(f"Worst deal on wings is for {max_awc} by purchasing {rowFinder(max_awc)['Quantity'].tolist()} wing(s)")
```

##### Value Deals
- The best deal on wings is for 111.2 cents/wing by purchasing 25, 50, or 125 wings
- Conversely, the worst deal on wings is for 114.0 cents/wing by purchasing 5 wings
